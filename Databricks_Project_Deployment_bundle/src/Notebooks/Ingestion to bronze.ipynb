```python
spark.sql(
    f"use catalog {catalog_name}"
)
spark.sql(
    "use schema bronze"
)
spark.sql(
    "create volume if not exists earthquake_data"
)
```

```python
import requests
import json
import datetime
url='https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_day.geojson'
response=requests.get(url)
if(response.status_code!=200):
  raise Exception(f"Error in fetching data from {url}")
data=response.json()
current_date=datetime.datetime.now().strftime('%Y-%m-%d')
dbutils.fs.put(
    f'/Volumes/{catalog_name}/bronze/earthquake_data/earthquake_data__{current_date}.json',
    json.dumps(data),
    overwrite=True
)
```
